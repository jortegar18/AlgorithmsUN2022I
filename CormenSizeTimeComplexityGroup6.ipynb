```python
import math

def nlogn(seg):
    n = 2
    maximo = seg*2.5*10**9
    nlogn = 0
    while(nlogn < maximo):
        n = maximo/math.log2(n)
        nlogn += (maximo-nlogn)/2
    return int(n)

print("Second: n <=", nlogn(1))
print("Minute: n <=", nlogn(60))
print("Hour: n <=", nlogn(3600))
print("Day: n <=", nlogn(86400))
print("Year: n <=", nlogn(31536000))
print("Century: n <=", nlogn(3153600000))
print("Milenium: n <=", nlogn(31536000000))
```

```python
import math

def nfact(seg):
    n = 0
    maximo = seg*2.5*10**9
    nfact = 1
    while(nfact <= maximo):
        n = n + 1
        nfact = math.factorial(n)
    return int(n-1)

print("Second: n <=", nfact(1))
print("Minute: n <=", nfact(60))
print("Hour: n <=", nfact(3600))
print("Day: n <=", nfact(86400))
print("Year: n <=", nfact(31536000))
print("Century: n <=", nfact(3153600000))
print("Millenium: n <=", nfact(31536000000))
```
